### Step 2: Load the Dataset

```python
# Load Dataset with Dask

data_path = "DataCoSupplyChainDataset.csv"  # Update this path if needed
dtypes = {
    "Customer Zipcode": "float64"
}
df = dd.read_csv(data_path, encoding='latin1', dtype=dtypes)
```

### Step 4: Data Preprocessing


```python
# Checking for Missing Values
missing_values = df.isnull().sum().compute()
print("Missing Values:\n", missing_values)
```

```python
# Filling Missing Values Using Forward Fill Method
df = df.fillna(method='ffill')
```

### Step 5: Exploratory Data Analysis (EDA)
#### Perform basic analysis to understand the dataset

### Delivery Performance Analysis - Evaluating Late Delivery Risk

```python
# Delivery Performance Analysis - Evaluating Late Delivery Risk

delivery_analysis = df.groupby("Late_delivery_risk").agg({
    "Order Id": "count",
    "Days for shipping (real)": "mean"
}).compute()
print("Delivery Performance Analysis:\n", delivery_analysis)
```

```python
# Plotting Late Delivery Risk Analysis

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.barplot(x=delivery_analysis.index, y=delivery_analysis["Order Id"], palette="viridis")
plt.xlabel("Late Delivery Risk")
plt.ylabel("Number of Orders")
plt.title("Late Delivery Risk Analysis")
plt.show()
```

### Sales and Profitability Analysis - Analyzing Sales Per Customer


```python
sales_per_customer = df.groupby("Order Customer Id").agg({
    "Sales": "sum",
    "Order Profit Per Order": "mean"
}).compute()
print("Sales and Profitability Analysis:\n", sales_per_customer.head())
```

```python
# Summary statistics for sales per customer
df['Sales per customer'].describe()
```

```python
#  Sales distribution plot
sns.histplot(df['Sales per customer'], bins=30, kde=True)
plt.title("Sales Per Customer Distribution")
plt.show()
```

### Inventory and Category Analysis - Category-wise Sales Analysis


```python
category_sales = df.groupby("Category Name").agg({
    "Sales": "sum",
    "Order Id": "count"
}).compute()
print("Category-wise Sales Analysis:\n", category_sales)
```

```python
# Plotting Category-wise Sales Analysis

plt.figure(figsize=(10, 6))
sns.barplot(x=category_sales.index, y=category_sales["Sales"], palette="coolwarm")
plt.xlabel("Category Name")
plt.ylabel("Total Sales")
plt.title("Category-wise Sales Analysis")
plt.xticks(rotation=45)
plt.show()
```

### City-wise Supply Chain Performance - Top Performing Cities Based on Sales


```python
# City-wise Supply Chain Performance - Top Performing Cities Based on Sales
top_cities = df.groupby("Order City").agg({
    "Sales": "sum"
}).compute().sort_values(by="Sales", ascending=False)
print("Top Performing Cities Based on Sales:\n", top_cities.head(10))
```

```python
# Plotting Top Performing Cities Based on Sales

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.index[:10], y=top_cities["Sales"].head(10), palette="magma")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.title("Top Performing Cities Based on Sales")
plt.xticks(rotation=45)
plt.show()
```
